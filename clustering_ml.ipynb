```python
imagearr=image.flatten()
```

```python
imagearr.shape
```

```python
image2arr=imagearr.reshape(786432,1)
```

```python
image2arr.shape
```

```python
from sklearn.cluster import KMeans
```

```python
km=KMeans(n_clusters=5,init='k-means++',random_state=33)
```

```python
comp_image=km.fit(image2arr)
```

```python
type(comp_image)
```

```python
clusters=np.asarray(km.cluster_centers_)
labels = np.asarray(km.labels_ )  
comp_image=labels.reshape(768, 1024)
```
